```python
churn_master_df = pd.read_csv("churn_data.csv")
```

```python
churn_df = churn_df.replace(r'^\s*$', np.nan, regex = True)
```

```python
from autoviz.AutoViz_Class import AutoViz_Class
AV = AutoViz_Class()
```

```python
dft = AV.AutoViz("churn_data.csv",sep=",",depVar="",dfte=None,header=1,verbose=1,lowess=False,
                 chart_format='svg',max_rows_analyzed=7500,max_cols_analyzed=50)
```

```python
dft = AV.AutoViz("churn_data.csv",sep=",",depVar="Churn",dfte=None,header=0,verbose=0,lowess=False,
                 chart_format='svg',max_rows_analyzed=7500,max_cols_analyzed=50)
```
